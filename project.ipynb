```python
import matplotlib.pyplot as plt
import numpy as np

def cal_a(k):
    sum_real = 0
    sum_img = 0
    
    for n in range(22):
        sum_real += (0.75) ** n * np.cos(k * (np.pi / 11) * n)
        sum_img += (0.75) ** n * np.sin(k * (np.pi / 11) * n)
        
    real = (1 / 22) * sum_real
    img = (1 / 22) * sum_img
    return real, img

for k in range(22):
    real_part_a, img_part_a = cal_a(k)
    x_real += real_part_a * np.cos(k * (np.pi / 11) * n)
    x_img += img_part_a * np.sin(k * (np.pi / 11) * n)

plt.figure(figsize=(12, 12), dpi=100)

plt.subplot(2, 1, 1)
plt.stem(n, x_real, use_line_collection = True)
plt.ylabel('real(x[n])')

plt.subplot(2, 1, 2)
plt.stem(n, x_img, use_line_collection = True)
plt.xlabel('n')
plt.ylabel('img(x[n])')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    if t == 0:
        return 1 / np.pi
    return np.sin(np.pi * t)/(np.pi * t)

a = 7
b = 9
x = np.linspace(-10, 10, 1000)
y1 = [f(t) for t in x]
y2 = [f(a * t) for t in x]
y3 = [f(a * t + b) for t in x]

plt.figure(figsize=(12, 15), dpi=100)

plt.subplot(3, 1, 1)
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('f(x)')

plt.subplot(3, 1, 2)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('f(x)')

plt.subplot(3, 1, 3)
plt.plot(x, y3)
plt.xlabel('x')
plt.ylabel('f(x)')

plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

b = 3

def f(t):
    if t < 0:
        return 0
    elif t <= b:
        return 1 - np.exp(-t)
    else:
        return -np.exp(-t) + np.exp(-t + b)

t1 = np.linspace(-10, 10, 1000)
y1 = [f(t) for t in t1]
    
t2 = np.linspace(0, b, 1000)
y2 = [f(t) for t in t2]

t3 = np.linspace(b, 20, 1000)
y3 = [f(t) for t in t3]

plt.figure(figsize=(12, 6), dpi=100)
plt.plot(t1, y1, label='0')
plt.plot(t2, y2, label='1 - e ^ (-t)')
plt.plot(t3, y3, label='-e ^ (-t) + e ^ (-t + b)')
plt.legend(loc='upper right')

plt.xlabel('t')
plt.ylabel('f(t)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# part a
w = np.linspace(-20, 20, 1000)
H = 4 / (4 + w ** 2)

# part b
t1 = np.linspace(-20, 20, 1000)
xs = np.cos(0.1 * t1)
xn = 0.3 * np.cos(10 * t1)
x_total = xs + xn

# part c
y = 4 * np.cos(0.1*t1) / 4.01


plt.figure(figsize=(12, 20), dpi=100)

plt.subplot(4, 1, 1)
plt.plot(w, H)
plt.xlabel('w')
plt.ylabel('H(jw)')
plt.title('part a')

plt.subplot(4, 1, 2)
plt.plot(t1, x_total)
plt.xlabel('t')
plt.ylabel('X total')
plt.title('part b')

plt.subplot(4, 1, 3)
plt.plot(t1, y)
plt.xlabel('t')
plt.ylabel('y')
plt.title('part c')

plt.subplot(4, 1, 4)
plt.plot(t1, x_total, label='x total')
plt.plot(t1, y, label='y')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('part d')
plt.legend()

plt.subplots_adjust(hspace=0.5)
plt.show()
```
